```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True) #drop first and last column
```

```python
df.head()
```

# **Train Test Split**

syntax: df.iloc[rows, columns] , iloc means index-based selection.

**df.iloc[ : , 1 : ] ----->  all rows , columns 1 → end**

**df.iloc[:, 0]** ----> all rows,column 0 , Column 0 is the diagnosis column.

**test_size = 0.2** -----> 20% → test set ,80% → training set


```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

#**Scaling**

```python
scaler = StandardScaler() #This creates the scaler object.Think of it as a tool that will learn: feature means,feature standard deviations.But nothing happens yet.

X_train = scaler.fit_transform(X_train) #Two things happen here.
#Step 1:— fit() The scaler learns statistics from the training data.For each feature it calculates:mean standard deviation
#Step 2:— transform() Then it scales the training data using:(𝑥−𝑚𝑒𝑎𝑛)/𝑠𝑡𝑑 So fit_transform() means: learn statistics + scale the data

X_test = scaler.transform(X_test)
#We DO NOT use fit_transform here.We only use: transform() Why? Because we must use the same statistics learned from training data.
#If we ran fit on the test data, the model would accidentally peek at the test set, which is cheating.
```

```python
X_train
```

```python
y_train #notice it gives strings and nn cant work on that , that is why we will have to transform these into numbers through label encoding
```

#**Label Encoding**

```python
encoder = LabelEncoder() #This creates a label encoder object.It will learn how to map labels: B → 0 , M → 1


y_train = encoder.fit_transform(y_train) #Two things happen:
#Step 1:— fit()  The encoder looks at the training labels and finds unique classes: [B, M] It assigns numbers: B → 0 M → 1
#Step 2:— transform()  It converts the labels. Example:  [M, B, B, M] becomes [1, 0, 0, 1]


y_test = encoder.transform(y_test) #we must use the same mapping learned from the training set.
#If we ran fit on the test set, we might accidentally create a different mapping, which would break the model.
```

```python
y_train
```

#**Numpy Arrays to PyTorch Tensors**

```python
X_train_tensor = torch.from_numpy(X_train.astype(np.float32))
X_test_tensor = torch.from_numpy(X_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
```

#**Definig The Model**

```python
import torch.nn as nn


class MySimpleNN(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):

    out = self.linear(features)
    out = self.sigmoid(out)

    return out
```

#Important Parameters

```python
learning_rate = 0.1
epochs = 25
```

```python
# define loss function
loss_function = nn.BCELoss() #binaryCrossEntropy
```

#**Training Pipline**

```python
# create model
model = MySimpleNN(X_train_tensor.shape[1])

# define optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate) #The optimizer updates the weights of the model using gradients.  new_weight = old_weight - learning_rate * gradient

# define loop
for epoch in range(epochs):

  # forward pass
  y_pred = model(X_train_tensor)

  # loss calculate
  loss = loss_function(y_pred, y_train_tensor.view(-1,1))

  # clear gradients
  optimizer.zero_grad()

  # backward pass
  loss.backward()

  # parameters update
  optimizer.step()  #Now the optimizer updates the weights.

  # print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

#**Evaluation**

```python
# model evaluation
with torch.no_grad(): #Don't build the computation graph , Don't store gradients
  y_pred = model.forward(X_test_tensor) #This runs a forward pass using the test data.
  y_pred = (y_pred > 0.5).float()  # > 0.5 → class 1 , ≤ 0.5 → class 0
  accuracy = (y_pred == y_test_tensor).float().mean() #This calculates how many predictions were correct.
  print(f'Accuracy: {accuracy.item()}')
```
