###6. Inplace Operation

PyTorch tensors normally create new memory when you change them. In-place operations skip that. They modify the existing tensor directly instead of creating another one.

```python
m = torch.tensor([1,2,3])
n = m + 1

print(m)
print(n)

#What happened?
#PyTorch created a new tensor for b. a stayed untouched.
```

```python
a = torch.tensor([1,2,3])
a.add_(1)  #Instead of creating new memory, PyTorch just edits the existing tensor:
```

```python
a
```

```python
torch.relu(m)
```

```python
m.relu_()
```

###7. Copying a Tensor

```python
a = torch.randint(0,10,(2,3))
a
```

```python
b = a
```

```python
b
```

```python
a[0][0]=0
a
```

```python
b  # notice : b also changes but we dont want that , so another way is to clone it
```

```python
print(id(a))
print(id(b))
```

```python
b = a.clone()
```

```python
print(a)
print(b)
```

```python
a[0][0]=2
a
```

```python
print(a)
print(b)

print("id of a ",id(a))
print("id of b",id(b))

# now b remains as it was , while a changes
```
