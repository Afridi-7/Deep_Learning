```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6)) #plot a window , Size = 10 inches wide, 6 inches tall.
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y):

    X = np.insert(X,0,1,axis=1) #inserting a column for bias.. now it looks like ([1, x1, x2])
    weights = np.ones(X.shape[1]) #initialize weights , All weights start as 1.
    lr = 0.1

    for i in range(1000):
        j = np.random.randint(0, X.shape[0]) #selecting random point to start the algorithm
        y_hat = step(np.dot(X[j],weights))  #actually dot product of weights and inputs : w0x0 + w1x1 + w2x2
        weights = weights + lr*(y[j]-y_hat)*X[j]

    return weights[0],weights[1:] #weights[0] → the bias (intercept)  weights[1:] → the rest of the weights (coefficients)
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_ , coef_ = perceptron(X,y) #tuple unpacking
```

```python
print(coef_) #weights
print(intercept_) #bias
```

```python
m = -(coef_[0]/coef_[1]) #Slope of the decision boundary.
b = -(intercept_/coef_[1]) #Y-intercept of the decision boundary.
```

```python
x_input = np.linspace(-3,3,100) #p.linspace(start, stop, number_of_points) It means:Create 100 evenly spaced numbers between -3 and 3.
y_input = m*x_input + b   #line equation : y = mx + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)  #Limits the vertical axis.Means:Only show y-values between -3 and 2.This prevents the graph from stretching awkwardly.
```
