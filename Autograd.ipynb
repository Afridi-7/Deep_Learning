# 3:

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
```

```python
w = torch.tensor(1.0 , requires_grad=True)
b = torch.tensor(0.0 , requires_grad=True)
```

```python
# Binary Cross-Entropy Loss for scalar
def binary_cross_entropy_loss(prediction, target):
    epsilon = 1e-8  # To prevent log(0)
    prediction = torch.clamp(prediction, epsilon, 1 - epsilon)
    return -(target * torch.log(prediction) + (1 - target) * torch.log(1 - prediction))
```

```python
z = w*x + b ## Weighted sum (linear part)
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
# Compute binary cross-entropy loss
loss = binary_cross_entropy_loss(y_pred, y)
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```

# 5: Disable Gradient Tracking

```python
x = torch.tensor(2.0, requires_grad=True)
x
```

```python
y = x ** 2
y
```

```python
y.backward()
```

```python
x.grad  # it increases every time you run , so to avoid it , you must disable the gradient tracking by following method
```

```python
# option 1 - requires_grad_(False)
x.requires_grad_(False)
```

```python
y = x ** 2

y.backward() # wont run , because gradient has been disabled
```

.

```python
# option 2 - detach()
x = torch.tensor(2.0, requires_grad=True)
x
```

```python
z = x.detach() # you create a copy of x which doesnt have gradient , so you use it where you dont need gradient to calculate
z
```

```python
y = x ** 2
y
```

```python
y1 = z ** 2
y1
```

```python
y.backward()
```

```python
y1.backward() # error because y1 has z which no longer carries gradient
```
